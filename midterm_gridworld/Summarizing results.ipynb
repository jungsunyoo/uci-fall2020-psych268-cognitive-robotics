```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
episode = open("episodeLen.txt")
episode_data = episode.read().splitlines()
episode_data = np.array(episode_data, dtype=float)
# print(episode_data)
plt.title('Episode length over time')
plt.plot(range(len(episode_data)),episode_data, 'b-')
```

```python
reward = open("isRewarded.txt")
reward_data = reward.read().splitlines()
reward_data = np.array(reward_data, dtype=float)
# print(episode_data)
plt.title('Cumulate sum of success over episodes')
plt.plot(np.arange(len(reward_data)),np.cumsum(reward_data))
```
